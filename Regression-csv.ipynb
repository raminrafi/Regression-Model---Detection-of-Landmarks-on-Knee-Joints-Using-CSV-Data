# Data Preprocessing

```python
PATH = "/Users/apple/Desktop/Development/freelance/upwork/brad-miles/Landmarks-Detection/csvs/"
files = glob.glob(PATH + '/*.csv', recursive=True)
```

```python
#defining known and unknown landmarks
landmarks = [
    'PCLOrigin',
    'femoralCenter',
    'greaterTrochanter',
    'hipCenter',
    'lateralCondyle',
    'lateralEpicondyle',
    'lateralPosteriorCondyle',
    'medialCondyle',
    'medialEpicondyle',
    'medialPosteriorCondyle',
    'medialSulcus',
    'midfemurCenter',
    'whitesideReference'
]

known_landmarks = [
    'femoralCenter',
    'greaterTrochanter',
    'hipCenter',
    'lateralCondyle',
    'lateralPosteriorCondyle',
    'medialCondyle',
    'medialPosteriorCondyle',
    'midfemurCenter'
]

unknown_landmarks = [
    'PCLOrigin',
    'lateralCondyle',
    'medialCondyle',
    'medialSulcus',
    'whitesideReference'
]
```

```python
df = pd.DataFrame()
for file in files:
    df1 = pd.read_csv(file).transpose()
    temp = []
    for item in df1.iloc[0]:
        temp.append(item)
    df1.columns = temp
    df = df.append(df1[1:4])
```

```python
df.head()
```

```python
df_x = df[known_landmarks]
df_y = df[unknown_landmarks]
```

```python
df_x.head()
```

```python
df_y.head()
```

```python
new_known_landmarks = []
for item in df_x.columns:
    new_known_landmarks.append(item + "_x")
    new_known_landmarks.append(item+"_y")
    new_known_landmarks.append(item+"_z")
    
new_df = []
for i in range(0, len(df_x), 3):
    row = [] 
    temp = df_x.iloc(0)
    for item in known_landmarks:
        row.append(temp[i][item])
        row.append(temp[i+1][item])
        row.append(temp[i+2][item])
    new_df.append(row)
new_df_x = pd.DataFrame(new_df)
new_df_x.columns = new_known_landmarks
```

```python
new_df_x.head()
```

```python
new_unknown_landmarks = []
for item in unknown_landmarks:
    new_unknown_landmarks.append(item + "_x")
    new_unknown_landmarks.append(item+"_y")
    new_unknown_landmarks.append(item+"_z")
    
new_df = []
for i in range(0, len(df_y), 3):
    row = [] 
    temp = df_y.iloc(0)
    for item in unknown_landmarks:
        row.append(temp[i][item])
        row.append(temp[i+1][item])
        row.append(temp[i+2][item])
    new_df.append(row)
    
new_df_y = pd.DataFrame(new_df)
new_df_y.columns = new_unknown_landmarks
```

```python
new_df_y["PCLOrigin_x"].head()
```

```python
new_df_x.head()
```

# Model Training

```python
from sklearn import linear_model
```

```python
lm_x = linear_model.LinearRegression()
model_x = lm_x.fit(new_df_x,new_df_y["PCLOrigin_x"])
predictions_x = lm_x.predict(new_df_x)
print("PCLOrigin_x : ", lm_x.score(new_df_x,new_df_y["PCLOrigin_x"]))
```

```python
lm_y = linear_model.LinearRegression()
model_y = lm_y.fit(new_df_x,new_df_y["PCLOrigin_y"])
predictions_y = lm_y.predict(new_df_x)
print("PCLOrigin_y : ", lm_y.score(new_df_x,new_df_y["PCLOrigin_y"]))
```

```python
lm_z = linear_model.LinearRegression()
model_z = lm_z.fit(new_df_x,new_df_y["PCLOrigin_z"])
predictions_z = lm_z.predict(new_df_x)
print("PCLOrigin_z : ", lm_z.score(new_df_x,new_df_y["PCLOrigin_z"]))
```
