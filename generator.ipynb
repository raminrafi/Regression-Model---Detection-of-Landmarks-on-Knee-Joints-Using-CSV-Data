```python
import numpy as np
import copy

output_path = '/code/'

marks = [
    'PCLOrigin',
    'lateralCondyle',
    'medialCondyle',
    'medialSulcus',
    'whitesideReference'
]
```

```python
def generat_data(name = 'medialSulcus',index = 1):
    print(index)
    image = np.load(output_path + name +'/images/' + '{}.npy'.format(index))
    label = np.load(output_path + name +'/label/' + '{}.npy'.format(index), allow_pickle=True)
    mark = np.load(output_path + name +'/label/' + '{}-2.npy'.format(index))

    np.save(output_path + name +'/images/' + '{}-{}-{}-{}.npy'.format(index, 0,1,2), image.transpose(0,1,2))
    np.save(output_path + name +'/images/' + '{}-{}-{}-{}.npy'.format(index, 0,2,1), image.transpose(0,2,1))
    np.save(output_path + name +'/images/' + '{}-{}-{}-{}.npy'.format(index, 1,0,2), image.transpose(1,0,2))
    np.save(output_path + name +'/images/' + '{}-{}-{}-{}.npy'.format(index, 1,2,0), image.transpose(1,2,0))
    np.save(output_path + name +'/images/' + '{}-{}-{}-{}.npy'.format(index, 2,0,1), image.transpose(2,0,1))
    np.save(output_path + name +'/images/' + '{}-{}-{}-{}.npy'.format(index, 2,1,0), image.transpose(2,1,0))

    np.save(output_path + name +'/label/' + '{}-{}-{}-{}.npy'.format(index, 0,1,2), mark.transpose(0,1,2))
    np.save(output_path + name +'/label/' + '{}-{}-{}-{}.npy'.format(index, 0,2,1), mark.transpose(0,2,1))
    np.save(output_path + name +'/label/' + '{}-{}-{}-{}.npy'.format(index, 1,0,2), mark.transpose(1,0,2))
    np.save(output_path + name +'/label/' + '{}-{}-{}-{}.npy'.format(index, 1,2,0), mark.transpose(1,2,0))
    np.save(output_path + name +'/label/' + '{}-{}-{}-{}.npy'.format(index, 2,0,1), mark.transpose(2,0,1))
    np.save(output_path + name +'/label/' + '{}-{}-{}-{}.npy'.format(index, 2,1,0), mark.transpose(2,1,0))

    label1 = copy.deepcopy(label)
    label2 = copy.deepcopy(label)
    label2[5], label2[4] = label2[4], label2[5]
    label3 = copy.deepcopy(label)
    label3[3], label3[4] = label3[4], label3[3]
    label4 = copy.deepcopy(label)
    label4[3], label4[4] = label4[4], label4[3]
    label4[4], label4[5] = label4[5], label4[4]
    label5 = copy.deepcopy(label)
    label5[4], label5[5] = label5[5], label5[4]
    label5[3], label5[4] = label5[4], label5[3]
    label6 = copy.deepcopy(label)
    label6[3], label6[5] = label6[5], label6[3]

    np.save(output_path + name +'/label/' + '{}-{}-{}-{}.npy'.format(index, 0,1,2), label1)
    np.save(output_path + name +'/label/' + '{}-{}-{}-{}.npy'.format(index, 0,2,1), label2)
    np.save(output_path + name +'/label/' + '{}-{}-{}-{}.npy'.format(index, 1,0,2), label3)
    np.save(output_path + name +'/label/' + '{}-{}-{}-{}.npy'.format(index, 1,2,0), label4)
    np.save(output_path + name +'/label/' + '{}-{}-{}-{}.npy'.format(index, 2,0,1), label5)
    np.save(output_path + name +'/label/' + '{}-{}-{}-{}.npy'.format(index, 2,1,0), label6)
```

```python
for index in range(14, 40):
    for name in marks:
        generat_data(name, index)
```
