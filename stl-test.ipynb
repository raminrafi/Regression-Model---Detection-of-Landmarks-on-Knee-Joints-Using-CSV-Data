```python
from stl.mesh import Mesh
```

```python
path = "../Landmarks-Detection/Dataset/Pat1/POD_JR_9060K_Patella.stl"

# Read the STL using numpy-stl
mesh = Mesh.from_file(path)
```

```python
import plotly.graph_objects as go
import numpy as np

pts = np.loadtxt(np.DataSource().open('/Users/apple/Desktop/Development/freelance/upwork/brad-miles/Landmarks-Detection/out.txt'))

```

```python
x, y, z = pts.T

fig = go.Figure(data=[go.Mesh3d(x=x, y=y, z=z,
                   alphahull=5,
                   opacity=0.4,
                   color='light blue')])
fig.show()
```

```python
new_text = ""
for item in mesh:
    new_text = new_text+ "{} {} {}\n".format(str(item[0]), str(item[1]), str(item[2]))
```

```python
with open("out.txt", "w+") as f:
    f.write(new_text)
```
